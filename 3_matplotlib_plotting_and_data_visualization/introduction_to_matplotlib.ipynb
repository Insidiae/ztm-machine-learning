## Three Methods on Creating Matplotlib Plots

### 1st Method

````python
fig = plt.figure() # Creates a figure
ax = fig.add_subplot() # Adds some axes
plt.show() # Displays the plot
````

### 2nd Method

````python
fig = plt.figure() # Creates a figure
ax = fig.add_axes([1, 1, 1, 1]) # Adds some axes
ax.plot(x, y) # Adds some data
plt.show() # Displays the plot
````

### 3rd Method (recommended)

````python
fig, ax = plt.subplots() # Creates a figure with axes
ax.plot(x, y); # Adds some data
````

## Creating Multiple Plots using Subplots

### 1st Method

````python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))

# Plot to each different axis
np.random.seed(69)
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
````

### 2nd Method

````python
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))

# Plot to each different axis
np.random.seed(69)
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000));
````

## pyplot vs OO method: Which one should you use?

- When doing some quick plots and visualization, you might want to use the pyplot method.
- For more complex visualizations, it might be better to use the OO method instead.

Let's take one more look at the heart disease dataset:

````python
heart_disease.head()
````

````python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
````

````python
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target");
````

We've been using the pyplot method for these past few graphs, but now we're running into some issues when we try to plot more complex graphs. Let's try the OO method now and make the subplots more presentable.

### OO method mixed with pyplot method

````python
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax)

ax.set_xlim([45, 100]);
````

### OO method from scratch

````python
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line to represent avg cholesterol levels
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
````

Now let's try doing subplots using the OO Method.

First, let's have another look at the over_50 dataset:

````python
over_50.head()
````

````python
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)

# Add data to 1st subplot
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

# Customize 1st subplot
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")

# Add legend to 1st subplot
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a meanline to 1st subplot
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")

# Add data to 2nd subplot
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

# Customize 2nd subplot
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Add legend to 2nd subplot
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline to 2nd subplot
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")

# Add title to whole figure
fig.suptitle("Heart Disease Analysis",
             fontsize=16,
             fontweight="bold");
````

### Choosing a Color Map

We can actually change the colors displayed on the grid by setting a [color map](https://matplotlib.org/stable/tutorials/colors/colormaps.html).

Let's try customizing one of our earlier plots with these.

````python
# Set the style, if you haven't already
# plt.style.use("dark_background")

# Initialize the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="coolwarm")

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line to represent avg cholesterol levels
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
````

### Cuztomizing the x and y Axis Limits

````python
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)

# Add data to 1st subplot
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="coolwarm")

# Customize 1st subplot
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")

# Customize axis limits for 1st subplot
ax0.set_xlim([50, 80])

# Add legend to 1st subplot
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a meanline to 1st subplot
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")

# Add data to 2nd subplot
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="coolwarm")

# Customize 2nd subplot
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Customize axis limits for 2nd subplot
ax1.set_ylim([60, 200])

# Add legend to 2nd subplot
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline to 2nd subplot
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")

# Add title to whole figure
fig.suptitle("Heart Disease Analysis",
             fontsize=16,
             fontweight="bold");
````
