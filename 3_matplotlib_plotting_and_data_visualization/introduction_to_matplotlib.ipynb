## Three Methods on Creating Matplotlib Plots

### 1st Method

````python
fig = plt.figure() # Creates a figure
ax = fig.add_subplot() # Adds some axes
plt.show() # Displays the plot
````

### 2nd Method

````python
fig = plt.figure() # Creates a figure
ax = fig.add_axes([1, 1, 1, 1]) # Adds some axes
ax.plot(x, y) # Adds some data
plt.show() # Displays the plot
````

### 3rd Method (recommended)

````python
fig, ax = plt.subplots() # Creates a figure with axes
ax.plot(x, y); # Adds some data
````

## Matplotlib Example Workflow

### 0. Import Matplotlib and get it ready for plotting in Jupyter

```py
%matplotlib inline
import matplotlib.pyplot as plt
```

### 1. Prepare data

````python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
````

### 2. Setup plot using the OO approach

````python
fig, ax = plt.subplots(figsize=(10, 10))
````

### 3. Plot the Data

````python
ax.plot(x, y)
````

### 4. Customize the Plot

````python
ax.set(title="Sample Plot",
       xlabel="x-axis",
       ylabel="y-axis")
````

### 5. Save and Show (you save the whole figure)

````python
fig.savefig("./img/sample_plot.png")
````

## Making Figures with NumPy Arrays

````python
x = np.linspace(0, 10, 100)
x
````

````python
# Create a line plot using the data
fig, ax = plt.subplots()
ax.plot(x, x ** 2);
````

### Scatter Plot

````python
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
````

````python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
````

### Bar Plot

````python
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store",
       ylabel="Price ($)");
````

````python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title="Dan's Nut Butter Store",
       xlabel="Price ($)");
````

### Histogram

````python
np.random.seed(69)
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
````

## Creating Multiple Plots using Subplots

### 1st Method

````python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))

# Plot to each different axis
np.random.seed(69)
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
````

### 2nd Method

````python
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))

# Plot to each different axis
np.random.seed(69)
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000));
````
